```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("titanic.csv")
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["Age"].fillna(30,inplace=True)
```

```python
df["Cabin"].fillna("F16",inplace=True)
```

```python
df["Embarked"].fillna("S",inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Sex"].value_counts()
```

```python
survival_rates=df.groupby("Pclass")["Survived"].mean()
```

```python
survival_rates
```

```python
survival_rates.plot(kind='bar', color='skyblue')
plt.title('Hayatta Kalma Oranı (Pclass\'a Göre)')
plt.xlabel('Pclass')
plt.ylabel('Hayatta Kalma Oranı')
plt.show()
```
